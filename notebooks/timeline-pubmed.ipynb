```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
os.listdir('/home/amansinha/Downloads/pubmed/')
```

```python
for i in range(1,19):
    filename = f'/home/amansinha/Downloads/pubmed/PubMed_Timeline_Results_by_Year ({i}).csv'
    with open(filename) as f:
        first_line = f.readline().split(':')[-1].strip()
        
    df = pd.read_csv(filename, skiprows=1)
    size = (15,10) 
    if (len(df)>65):
        size = (50,15)
    ax = df.plot.bar(x='Year', y='Count', rot=90, title= first_line, figsize=size)
    for p in ax.patches:
        ax.annotate(str(p.get_height()), (p.get_x(), p.get_height()* 1.02), rotation=90)
    plt.legend(loc='upper left')
    plt.show()
    
    
```
